```python
import nltk
```

```python
from nltk.tokenize import wordpunct_tokenize 
```

```python
tokenized_text = wordpunct_tokenize("This is a test.")
tokenized_text
```

```python
def getVectorsOf(model, text):
    vectors = []
    for token in wordpunct_tokenize(text):
        try:
            vectors.append(model[token])
        except:
            pass
    return vectors
```

```python
import numpy as np
np.array(getVectorsOf(model, "This is a test.")).mean(axis=0)
```
